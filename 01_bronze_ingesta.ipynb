```python
# ============================================================
# NOTEBOOK 01 — BRONZE: INGESTA Y VERIFICACIÓN
# Vermont School - Sistema de Alerta Temprana 2025-26
# ============================================================
import os

BRONZE = "/Volumes/workspace/vermont/bronze"

archivos = os.listdir(BRONZE)
print(f"Archivos en bronze: {len(archivos)}")
for a in sorted(archivos):
    size = os.path.getsize(f"{BRONZE}/{a}")
    print(f"  {a:50} {size:>10,} bytes")
```

```python
# Verificar lectura de cada archivo
import pandas as pd
from io import StringIO

config = {
    "notas":      ("report.xls",                      "html"),
    "asistencia": ("exported.xls",                    "html"),
    "f1":         ("Consolidado_seguimiento.xls",      "csv"),
    "f2": ("Consolidado_seguimiento (1).xls", "csv"),
}

for nombre, (archivo, tipo) in config.items():
    ruta = f"{BRONZE}/{archivo}"
    try:
        if tipo == "html":
            with open(ruta, "r", encoding="utf-8", errors="replace") as f:
                df = pd.read_html(StringIO(f.read()))[0]
        else:
            df = pd.read_csv(ruta, sep=";", encoding="utf-8-sig")
        print(f"✓ {nombre:15} → {df.shape[0]:4} filas × {df.shape[1]:2} cols")
    except Exception as e:
        print(f"✗ {nombre:15} → ERROR: {e}")
```
